Numerical data needs StandardScaler, and categorical data needs OneHotEncoder.

````python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder

numeric_features = [
    'LIMIT_BAL', 'AGE', 'BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 
    'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6', 'PAY_AMT1', 'PAY_AMT2', 
    'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6', 'AVG_BILL_AMT', 'BILL_AMT_CHANGE'
]

categorical_features = ['SEX', 'EDUCATION', 'MARRIAGE']

passthrough_features = ['PAY_0', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6', 'PAY_TREND']

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features),
        ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), categorical_features),
        ('pass', 'passthrough', passthrough_features)
    ]
)

X_train_transformed = preprocessor.fit_transform(X_train_engineered)
print(f"Original shape: {X_train_engineered.shape}")
print(f"Transformed shape: {X_train_transformed.shape}")
````

````python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV, cross_validate

lr_pipe = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression(max_iter=1000, class_weight='balanced', random_state=123))
])
````

````python
param_grid = {
    'classifier__C': [0.001, 0.01, 0.1, 1, 10, 100]
}

grid_search = GridSearchCV(
    lr_pipe, param_grid, cv=5, scoring='f1', n_jobs=-1, return_train_score=True
)
grid_search.fit(X_train_engineered, y_train)

best_C = grid_search.best_params_['classifier__C']
print(f"Best C value: {best_C}")
````

````python
cv_results = cross_validate(
    grid_search.best_estimator_, 
    X_train_engineered, y_train, 
    cv=5, 
    scoring=['f1', 'roc_auc', 'accuracy'],
    return_train_score=True
)

import pandas as pd
res_df = pd.DataFrame(cv_results)
print(res_df.mean())
````
